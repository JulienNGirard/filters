```python
def plotlowpass(R,C):
    omega=np.logspace(-5,5,200)
    H=1/(1+1j*R*C*omega)
    Hnorm=20*np.log10(np.abs(H))
    #print Hnorm
    t=plt.figure(figsize=(10,10))
    plt.plot(omega,Hnorm)
    plt.xscale('log')
    plt.ylim(-100,5)
    plt.show()
```

```python
from scipy.fftpack import fft
# Number of samplepoints
N = 1000
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(5 * 2.0*np.pi*x) + 0.5*np.sin(1000.0 * 2.0*np.pi*x)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(2*np.pi*xf, 2.0/N * np.abs(yf[0:N/2]))
plt.xscale('log')
plt.grid()
plt.show()
```
